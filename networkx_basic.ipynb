```python
import networkx as nx
```

```python
g = nx.Graph()
```

```python
g.add_node("A")
g.add_nodes_from(["B", "C", "D"])
```

```python
g.add_edge("A", "B")
g.add_edges_from([("A", "C"), ("B", "D")])
```

```python
print(g.nodes())
```

```python
print(g.edges())
```

```python
print(g.number_of_nodes())
```

```python
print(g.number_of_edges())
```

```python
import matplotlib.pyplot as plt

nx.draw(g, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```

```python
dg = nx.DiGraph()
dg.add_edge("A", "B")
```

```python
nx.draw(dg, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```

```python
wg = nx.Graph()
wg.add_edge("A", "B", weight=4)
wg.add_edge("A", "C", weight=2)
wg.add_edge("B", "C", weight=5)
wg.add_edge("B", "D", weight=10)
wg.add_edge("C", "D", weight=3)
```

```python
pos = nx.spring_layout(wg, seed=42)
nx.draw(wg, pos, with_labels=True, node_color='lightblue', font_weight='bold')
edge_labels = nx.get_edge_attributes(wg, 'weight')
nx.draw_networkx_edge_labels(wg, pos, edge_labels=edge_labels, font_color='red')
plt.show()
```

```python
n = nx.Graph()
```

```python
n.add_nodes_from(["A","B","C","D","E","F","G"])
```

```python
n.add_edges_from([("A","B"), ("B","C"), ("C","D"), ("A","D"), ("D","E"), ("B","F"), ("F","G"), ("C","G"), ("D","G")])
```

```python
nx.draw(n, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```

```python
print(nx.shortest_path(n, "A", "G"))
```

```python
print(dict(n.degree()))
```

```python
print(list(nx.connected_components(n)))
```

```python
print(nx.degree_centrality(n))
print(nx.betweenness_centrality(n))
```
